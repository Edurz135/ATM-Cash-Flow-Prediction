# Feature Engineering

```python
import pandas as pd
import seaborn as sns
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("AggregatedData.csv")
```

```python
data
```

```python
data.describe()
```

```python
data['Weekday'] = data['Weekday'].str.upper()
data
```

```python
data['Transaction Date'] = pd.to_datetime(data['Transaction Date']) 
data
```

```python
new_data = data.drop([ 'Amount withdrawn XYZ Card',
       'Amount withdrawn Other Card','No Of XYZ Card Withdrawals', 'No Of Other Card Withdrawals'], axis =1)
new_data
```

```python
bigstreet = new_data[new_data['ATM Name']=='Big Street ATM']
bigstreet
```

```python
weekly_SMA = bigstreet.iloc[:,3].rolling(window=7).mean()
```

```python
weekly_SMA
```

```python
new_bigstreet = pd.DataFrame(weekly_SMA)
```

```python
new_bigstreet['Prev week mean'] = weekly_SMA
new_bigstreet
```

```python
monthly_SMA = bigstreet.iloc[:,3].rolling(window=30).mean()
```

```python
new_bigstreet['Prev month mean'] = pd.DataFrame(monthly_SMA)
```

```python
new_bigstreet.head(40)
```

```python
yearly_SMA = bigstreet.iloc[:,3].rolling(window=365).mean()
```

```python
new_bigstreet['Prev year mean'] = pd.DataFrame(yearly_SMA)
```

```python
new_bigstreet = new_bigstreet.drop(['Total amount Withdrawn'], axis = 1)
```
